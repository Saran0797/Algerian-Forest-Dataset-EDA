```python
categorical_feature = [feature for feature in data.columns if data[feature].dtype == 'O']
categorical_feature
```

```python
numerical_feature = [feature for feature in data.columns if data[feature].dtype!='O']
numerical_feature
```

```python
box_plot = data[['Temperature', 'RH', 'Ws', 'Rain', 'FFMC', 'DMC', 'DC', 'ISI', 'BUI', 'FWI']]
```

```python
box_plot
```

```python
plt.figure(figsize=(15,15))
plt.suptitle('box plot of each numerical variables', fontsize=20, fontweight='bold')

for i in range(0, len(box_plot.columns)):
    plt.subplot(4,4,i+1)
    sns.boxplot(x=box_plot.columns[i], data=data)
    plt.xlabel(box_plot.columns[i])
    plt.tight_layout()
```

```python
plt.figure(figsize=(15,15))
plt.suptitle('distribution of each numerical variables', fontsize=20, fontweight='bold')

for i in range (0, len(box_plot.columns)):
    plt.subplot(4,4,i+1)
    sns.distplot(data[box_plot.columns[i]])
    plt.xlabel(box_plot.columns[i])
    plt.tight_layout()
```
